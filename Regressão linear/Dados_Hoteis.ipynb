```python
from sklearn.model_selection import train_test_split
```

```python
y = dados_hoteis['Preco']
x = dados_hoteis.drop(columns = 'Preco')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=230)
```

```python
df_train = pd.DataFrame(data=x_train)
df_train['Preco'] = y_train
```

```python
from statsmodels.formula.api import ols
```

```python
modelo_0 = ols('Preco ~ Estrelas', data = df_train).fit()
```

```python
y_predict = modelo_0.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```

```python
import statsmodels.api as sm
```

```python
x_train_const = sm.add_constant(x_train)
x_train_const.head()
```

```python
x_train.columns
```

```python
modelo_1 = sm.OLS(y_train, x_train_const[['const','Estrelas', 'ProximidadeTurismo', 'Capacidade']]).fit()
```

```python
modelo_2 = sm.OLS(y_train, x_train_const[['const','Estrelas', 'Capacidade']]).fit()
```

```python
modelo_3 = sm.OLS(y_train, x_train_const[['const','Estrelas', 'ProximidadeTurismo']]).fit()
```

```python
modelo_0.summary()
```

```python
modelo_1.summary()
```

```python
modelo_2.summary()
```

```python
modelo_3.summary()
```

### Comparando Modelos

```python
print('R²')
print(f'modelo 0: {modelo_0.rsquared}')
print(f'modelo 1: {modelo_1.rsquared}')
print(f'modelo 2: {modelo_2.rsquared}')
print(f'modelo 3: {modelo_3.rsquared}')
```

```python
print(f'modelo 0 - parametros: {len(modelo_0.params)}')
print(f'modelo 1 - parametros: {len(modelo_1.params)}')
print(f'modelo 2 - parametros: {len(modelo_2.params)}')
print(f'modelo 3 - parametros: {len(modelo_3.params)}')
```

```python
x_test_const = sm.add_constant(x_test)
x_test_modelo2 = sm.add_constant(x_test.drop(columns = 'ProximidadeTurismo'))
```

```python
y_predict_1 = modelo_1.predict(x_test_const)
y_predict_2 = modelo_2.predict(x_test_modelo2)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
erro_quadratico_medio_modelo1 = mean_squared_error(y_test, y_predict_1)
erro_medio_absoluto_modelo1 = mean_absolute_error(y_test, y_predict_1)
r2_1 = r2_score(y_test, y_predict_1)
```

```python
erro_quadratico_medio_modelo2 = mean_squared_error(y_test, y_predict_2)
erro_medio_absoluto_modelo2 = mean_absolute_error(y_test, y_predict_2)
r2_2 = r2_score(y_test, y_predict_2)
```

```python
print('R2')
print(f'modelo 1: {r2_1}')
print(f'modelo 2: {r2_2}')
print('Erro Quadratico Médio')
print(f'modelo 1: {erro_quadratico_medio_modelo1}')
print(f'modelo 2: {erro_quadratico_medio_modelo2}')
print('Erro Absoluto Médio')
print(f'modelo 1: {erro_medio_absoluto_modelo1}')
print(f'modelo 2: {erro_medio_absoluto_modelo2}')
```
